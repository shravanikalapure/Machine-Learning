```python
#Fill 'Married' with mode (categorical)
df["Married"] = df["Married"].fillna(df["Married"].mode()[0])

#Fill 'Married' with mode (categorical)
df["Dependents"] = df["Dependents"].fillna(df["Dependents"].mode()[0])

#Fill 'Married' with mode (categorical)
df["Self_Employed"] = df["Self_Employed"].fillna(df["Self_Employed"].mode()[0])
```

```python
#Detection of outliers with IQR Method
Q1 = df["ApplicantIncome"].quantile(0.25)
Q3 = df["ApplicantIncome"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_iqr = df[(df["ApplicantIncome"] < lower_bound) | (df["ApplicantIncome"] > upper_bound)]

print("Outliers detected by IQR: ")
outliers_iqr["ApplicantIncome"]
```

```python
#Oulier Detection with ZScore Method
from scipy.stats import zscore
df["Zscore"] = zscore(df["ApplicantIncome"])
df.sample(5)
```

```python
outliers_z = df[(df["Zscore"] < -3) | (df["Zscore"] > 3)]

print("Outliers detected by ZScore: ")
print(outliers_z)
```

```python
outlier_z = df[df["Zscore"].abs() > 3]

print("Outliers detected by ZScore: ")
print(outlier_z)
```

```python
#Dimensionality Reduction By PCA Method
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Drop non-numeric columns and handle missing values
X = df.select_dtypes(include=['int64', 'float64']).fillna(0)

# Standardize the numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA (reduce to 2 principal components for visualization)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Explained variance
print("Explained variance ratio by each component:", pca.explained_variance_ratio_)

# Scatter plot of the 2 principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='skyblue', s=50, edgecolor='k')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Projection of Loan Dataset (Numerical Features Only)")
plt.grid(True)
plt.show()
```
