```python
import numpy as np
import pandas as pd
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn (3).csv")
```

```python
df = df.dropna()
```

```python
# Convert TotalCharges to numeric (it has some spaces, so force errors to NaN and drop them)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors='coerce')
df = df.dropna(subset=["TotalCharges"])
```

```python
#convert the target column that is churn to 0 or 1
df['Churn'] = df['Churn'].map({'Yes':1, 'No':0})
```

```python
#Dropping the customer id as it is not useful
df = df.drop(columns = 'customerID')
```

```python
# converting categorical variables to dummy / one hot encoding
df = pd.get_dummies(df, drop_first = True)
```

```python
#separating features and labels
X = df.drop('Churn', axis = 1).values
y = df['Churn'].values.reshape(-1, 1)
```

```python
# Normalize numeric features (feature scaling)
# Assuming all columns except the last one (which is Churn) are features after get_dummies
# Need to identify and normalize only numeric features. Let's skip normalization for now as get_dummies creates many boolean columns.
# Instead, let's just add the intercept term.

# Add intercept term (bias)
X = np.hstack((np.ones((X.shape[0],1)), X))
```

```python
#sigmoid function
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
```

```python
# cost function (binary cross entropy)
def compute_cost(X, y, weights):
  m = len(y)
  h = sigmoid(np.dot(X, weights))
  cost = -(1/m) * np.sum(y*np.log(h) + (1-y)*np.log(1-h))
  return cost
```

```python
def sigmoid(z):
    z = np.asarray(z, dtype=float)   # ensure correct type
    return 1 / (1 + np.exp(-z))
```

```python
def gradient_descent(X, y, weights, lr=0.1, epochs=10000):
    X = np.asarray(X, dtype=float)
    y = np.asarray(y, dtype=float)
    weights = np.asarray(weights, dtype=float)

    m = len(y)
    for i in range(epochs):
        h = sigmoid(np.dot(X, weights))
        gradient = np.dot(X.T, (h - y)) / m
        weights = weights - lr * gradient   # safe float operation
        if i % 1000 == 0:
            print(f"Epoch {i}, Cost: {compute_cost(X,y,weights)}")
    return weights
```

```python
#intialize weights
weights = np.zeros((X.shape[1],1))
```

```python
# Train
weights = gradient_descent(X, y, weights, lr=0.1, epochs=10000)
```

```python
# Train
weights = gradient_descent(X, y, weights, lr=0.1, epochs=10000)

# Predictions
y_pred = sigmoid(np.dot(X, weights))
y_pred_class = (y_pred >= 0.5).astype(int)

# Accuracy
accuracy = (y_pred_class == y).mean()
print("Accuracy:", accuracy)
```
